Task 1: Working with Series:-


1. Create a Pandas Series from the following list:
data = [25, 30, 35, 40, 45]
2. Write a Python program to:

-->Assign custom indices ['A', 'B', 'C', 'D', 'E'] to the Series.

-->Display the first three elements of the Series.

-->Calculate the mean, median, and standard deviation of the Series.

```python
import pandas as pd
```

```python
data = [25, 30, 35, 40, 45]

pds=pd.Series(data)
print("Original series : ")
print(pds)

print("Series with index : ")
pds.index=['A', 'B', 'C', 'D', 'E']
print(pds)

print("Displaying the first three elements of the Series :")
print(pds.head(3))

print("Statistics :-")
print(f"Mean : {pds.mean()}")
print(f"Median : {pds.median()}")
print(f"Standard deviation : {pds.std()}")
```

Task 7: General :-


1. Download a sample dataset from Kaggle or use any public dataset.
2. Write a Python program to:

-->Load the dataset using Pandas.

-->Perform exploratory data analysis (EDA) by summarizing key statistics, checking for missing values, and visualizing data trends using Matplotlib or Seaborn.

-->Document your findings in Markdown or comments.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('hotel_booking.csv')

# Basic Info
print(df.shape)
print(df.info())
print(df.describe())

# Check for nulls
print(df.isnull().sum().sort_values(ascending=False))

# Fill or drop missing values (example: fill with median or drop)
df['children'].fillna(0, inplace=True)
df.dropna(subset=['country'], inplace=True)

# Univariate Analysis
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='hotel')
plt.title('Number of Bookings by Hotel Type')
plt.show()

sns.histplot(df['lead_time'], bins=30, kde=True)
plt.title('Distribution of Lead Time')
plt.show()

# Bivariate Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x='is_canceled', y='lead_time', data=df)
plt.title('Lead Time vs Booking Cancellation')
plt.show()

# Aggregation: Cancellation rate by hotel
cancel_rate = df.groupby('hotel')['is_canceled'].mean().reset_index()
sns.barplot(data=cancel_rate, x='hotel', y='is_canceled',hue="hotel")
plt.title('Cancellation Rate by Hotel')
plt.ylabel('Cancellation Rate')
plt.show()

# Heatmap for correlation
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()

# Country-wise booking heatmap (Top 10)
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Booking Countries')
plt.xlabel('Number of Bookings')
plt.show()
```
